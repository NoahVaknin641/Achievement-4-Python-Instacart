```python
#5
ords_prods_merge.groupby('loyalty_flag')['prices'].agg(['mean', 'min', 'max', 'median', 'sum'])
```

New customers spend the most on average followed by regular sustomers followed by loyal customers

```python
#6
mean_price_per_user = ords_prods_merge.groupby('user_id')['prices'].mean()
ords_prods_merge.loc[ords_prods_merge['user_id'].isin(mean_price_per_user[mean_price_per_user >= 10].index), 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['user_id'].isin(mean_price_per_user[mean_price_per_user < 10].index), 'spending_flag'] = 'Low spender'
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.groupby('spending_flag')['prices'].mean()
```

I calculated the mean price of each user. then assigned the spending flag, finally I checked the value count per flag

```python
# Step 1: Assign "Non-frequent customer" to users with a median > 20
ords_prods_merge.loc[ords_prods_merge['days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'

# Step 2: Assign "Regular customer" to users with a median > 10 and <= 20
ords_prods_merge.loc[(ords_prods_merge['days_since_prior_order'] > 10) & (ords_prods_merge['days_since_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'

# Step 3: Assign "Frequent customer" to users with a median <= 10
ords_prods_merge.loc[ords_prods_merge['days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'

# Step 4: Check the results
ords_prods_merge['order_frequency_flag'].value_counts(dropna=False)
```
