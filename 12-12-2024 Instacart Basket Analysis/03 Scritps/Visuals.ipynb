```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
path = r'C:\Users\nvak6\12-12-2024 Instacart Basket Analysis'
merged_data = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'merged_data.pkl'))
```

```python
np.random.seed(4)
dev = np.random.rand(len(merged_data)) <= 0.7
```

```python
big = merged_data[dev]
small = merged_data[~dev]
```

```python
df_small = small[['order_hour_of_day','prices']]
```

```python
line = sns.lineplot(data = df_small, x = 'order_hour_of_day', y = 'prices', color='#68bf43')
plt.xlabel("Hour of Day")
plt.ylabel("Average Price of Products Purchased")
plt.title("Average Price of Instacart Products Purchased by Hour of Day")
```

```python
df_2_small = small[['Age', 'n_dependants']]
```

```python
line_2 = sns.lineplot(data = df_2_small, x = 'Age', y = 'n_dependants', color='#68bf43')
plt.xlabel("Customers Age")
plt.ylabel("Number of Dependants")
plt.title("Dependants by Age Distribution")
```

```python
hist_order_hour = merged_data['order_hour_of_day'].hist(bins=50)
plt.xlabel('Hour of Day')  # x-axis label
plt.ylabel('Number of Orders')  # y-axis label
plt.title('Distribution of Orders by Hour of the Day')  # title
```

```python
import matplotlib.pyplot as plt

# Create the bar chart
bar_loyalty = merged_data['loyalty_flag'].value_counts().plot.bar(color=['red', 'blue', 'yellow'])

# Add labels and title
plt.xlabel('Loyalty Flag')  # x-axis label
plt.ylabel('Number of Customers')  # y-axis label
plt.title('Customer Distribution by Loyalty Flag')  # title

# Display the plot
plt.show()
```

```python
# Create the scatterplot
scatter = sns.scatterplot(x='Age', y='income', data=merged_data)
# Add labels and title
plt.xlabel('Age')  # x-axis label
plt.ylabel('Income')  # y-axis label
plt.title('Relationship Between Age and Income')  # title
```
