```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
path = r'C:\Users\nvak6\12-12-2024 Instacart Basket Analysis'
merged_data = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'merged_data.pkl'))
```

```python
#3 Histogram of order_hour_of_day
hist_order_hour = merged_data['order_hour_of_day'].hist(bins = 50)
```

```python
# this plots the hours with the highest frequency of orders. 50 bins so you can see each our without overlap
```

```python
# Bar plot of loyalty flag
bar_loyalty = merged_data['loyalty_flag'].value_counts().plot.bar(color = ['red', 'blue', 'yellow'])
```

```python
# Difference in price depending on hour of the day
np.random.seed(4)
dev = np.random.rand(len(merged_data)) <= 0.7
```

```python
big = merged_data[dev]
small = merged_data[~dev]
```

```python
df_small = small[['order_hour_of_day','prices']]
```

```python
sample_line  = sns.lineplot(data = df_small, x = 'order_hour_of_day', y = 'prices')
```

```python
sample_line_2  = sns.lineplot(data = df_2_small, x = 'Age', y = 'n_dependants')
```

```python
# scatter plot of age and income
scatter = sns.scatterplot(x = 'Age', y = 'income', data = merged_data)
```
